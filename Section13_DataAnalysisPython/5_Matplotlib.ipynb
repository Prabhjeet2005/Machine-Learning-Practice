## Basic Line Plot


```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]  # X Coordinates
y = [1, 4, 9, 16, 25, 36]  # Y Coordinates

## Create Line Plot
plt.plot(x, y)  # Plotting Graph
plt.xlabel("X Axis")  # X Axis Heading
plt.ylabel("Y Axis")  # Y Axis Heading
plt.title("Basic Line Plot")  # Graph Heading
plt.show()  # Graph Display
```

## Customized Basic Line Plot


```python
x = [1, 2, 3, 4, 5, 6]
y = [1, 4, 9, 16, 25, 36]

plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.title("Customised Line Graph")

plt.plot(x, y, color="green", linestyle="--", marker="o", linewidth="2", markersize=6)
plt.grid(True)  # Forms A Grid In Graph


# color 		=> Colors Line
# lineStyle 	=> How line is Displayed
# marker 		=> (x,y) coordinate Marker
# linewidth		=> Thickness of Line
# markersize 	=> Thickness of Marker
```

# Multiple Plots

### OUTPUT OF GRAPH LAYOUT:-

#### .subplot(nrows,ncols,Plot Display Position)


```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]

plt.figure(figsize=(9, 5))

plt.subplot(
    1, 2, 1
)  # OUTPUT OF GRAPH LAYOUT:-	subplot(nrows,ncols,Plot Display Position)
plt.plot(x, y1, color="green")
plt.title("Plot 1")

plt.subplot(1, 2, 2)
plt.plot(y1, x, color="red")
plt.title("Plot 2")
```

# Bar Graph


```python
categories = ["A", "B", "C", "D", "E"]
values = [10, 20, 15, 30, 5]

plt.bar(categories, values, color="Orange")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Graph")
```

## Histogram

- #### They Divide data into Bins and count no. of data points in Each Bin

## X-axis Width ??

### Width X-Axis = Range/No. of Bins


```python
data = [0, 2, 3, 4, 3, 7, 7, 9, 9, 9, 9, 10]

plt.figure(figsize=(9, 5))
plt.subplot(1, 2, 1)

# Main Histogram
plt.hist(data, bins=5, color="orange", edgecolor="black")
# Range = 10, No. Of Bins = 5
# X-Axis Width = Range/No. Of Bins = 10/5 = 2	(Therefore Each X-Axis Occupy 2-Width)

plt.subplot(1, 2, 2)
plt.plot(data)
```

# Pie Chart


```python
import matplotlib.pyplot as plt

labels = ["A", "B", "C", "D"]
sizes = [20, 10, 30, 40]
pieChartColors = ["lightgreen", "lightblue", "lightpink", "orange"]
explode = (0.2, 0, 0, 0)  # Move Out the First Slice

plt.title("Pie Chart")
plt.pie(
    sizes,
    labels=labels,
    explode=explode,
    colors=pieChartColors,
    autopct="%1.3f%%",  # Automatic Percentage
    shadow=True,
)
```

# Practical Application


```python
import pandas as pd

salesData_dataFrame = pd.read_csv("salesData.csv")
salesData_dataFrame.head(5)
```

```python
salesData_dataFrame.info()
```

```python
sales_product_group = salesData_dataFrame.groupby("Product Category")[
    "Total Revenue"
].sum()
sales_product_group
```

# Plotting Bar Graph On Sales Data
## Way1	
 ## dataFrame.plot()	-> Quick Analysis	Easy
## Way2	
 ## plt.bar()			-> More Control		Hard


```python
# Way 1

import matplotlib.pyplot as plt

sales_product_group.plot(kind="bar", color="orange", edgecolor="black")
```

```python
# Way2
# Plotting Bar Graph on Sales Data

plt.figure(figsize=(9, 5))
plt.bar(
    sales_product_group.index,
    sales_product_group.values,
    color="orange",
    edgecolor="black",
)
```
