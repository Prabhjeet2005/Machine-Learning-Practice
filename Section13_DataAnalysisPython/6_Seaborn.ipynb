# Seaborn

### Python Visualisation Library (Based on Matplotlib) Used For:-

- #### Attractive & informative statistical graphs
- #### Complex Visualisations


```python
import seaborn as sns
```

## Basic Plotting with Seaborn


```python
# many Datasets under .load_dataset()
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips.info()
```

# Scatterplot


```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("total_bill V/S tip Graph")
```

# Line Plot


```python
sns.lineplot(x="size", y="total_bill", data=tips)
plt.title("Size V/S total_bill Line Plot")
```

# Categorical Plots

- ### Categories would be sex, smoker, day, time


## Bar Plot

```python
# Day V/s total_bill Bar Plot
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("day V/S total_bill Bar Plot")
```

# Box Plot



```python
sns.boxplot(x="day", y="total_bill", data=tips)
```


# Violin Plot



```python
sns.violinplot(x="day", y="total_bill", data=tips)
```

# Histogram Plot

```python
sns.histplot(tips["total_bill"], bins=10, kde=True)
```

# kde Plot


```python
sns.kdeplot(tips["total_bill"], fill=True)
```

# Pair Plot


```python
sns.pairplot(tips)
```

# Heatmap


```python
# Heatmap
corr = tips[["total_bill", "size", "tip"]].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap="coolwarm")
```
