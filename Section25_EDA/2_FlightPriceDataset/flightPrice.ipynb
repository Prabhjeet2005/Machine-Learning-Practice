# Cleaning Date_of_Journey


```python
# Convert DD/MM/YYYY to Date, Month, Year
df["Date"] = df["Date_of_Journey"].str.split("/").str[0]
df["Month"] = df["Date_of_Journey"].str.split("/").str[1]
df["Year"] = df["Date_of_Journey"].str.split("/").str[2]
```

```python
df.info()  # Still Date,Month,Year in Object
# Convert it to int
```

```python
## Convert Object to int
df["Date"] = df["Date"].astype(int)
df["Month"] = df["Month"].astype(int)
df["Year"] = df["Year"].astype(int)

df.info()
```

```python
# Drop Date_of_Journey
df.drop("Date_of_Journey", axis=1, inplace=True)
df.head()
```

# Cleaning Arrival_Time


```python
# Removing Unnecessary 10 Jun or 22 Mar After Time
df["Arrival_Time"] = df["Arrival_Time"].apply(lambda x: x.split(" ")[0])
df["Arrival_Time"]
```

```python
# Dividing into Hour and Minute
df["Arrival_Hour"] = df["Arrival_Time"].str.split(":").str[0]
df["Arrival_Minute"] = df["Arrival_Time"].str.split(":").str[1]

df.head()
```

```python
df["Arrival_Hour"] = df["Arrival_Hour"].astype(int)
df["Arrival_Minute"] = df["Arrival_Minute"].astype(int)

df.info()
```

```python
# Dropping Arrival_Time
df.drop("Arrival_Time", axis=1, inplace=True)
```

```python
df.head()
```

# Cleaning Dep_Time

```python
df["Dep_Hour"] = df["Dep_Time"].str.split(":").str[0]
df["Dep_Minute"] = df["Dep_Time"].str.split(":").str[1]
```

```python
df["Dep_Hour"] = df["Dep_Hour"].astype(int)
df["Dep_Minute"] = df["Dep_Minute"].astype(int)
df.info()
```

```python
df.drop("Dep_Time", axis=1, inplace=True)
```

```python
df.head(1)
```

# Cleaning Total_Stops

```python
df["Total_Stops"].unique()
```

```python
df[df["Total_Stops"].isnull()]
```

```python
df["Total_Stops"].mode()
```

```python
# Map With Ordinal values and assign nan with mode value
df["Total_Stops"] = df["Total_Stops"].map(
    {"non-stop": 0, "2 stops": 2, "1 stop": 1, "3 stops": 3, np.nan: 0, "4 stops": 4}
)
```

```python
df[df["Total_Stops"].isnull()]
```

```python
df.head(3)
```

# Drop Route Feature

```python
df.drop("Route", axis=1, inplace=True)
df.head(1)
```

# Cleaning Duration Feature

```python
df["Duration_Hour"] = df["Duration"].str.split(" ").str[0].str.split("h").str[0]
df["Duration_Minute"] = df["Duration"].str.split(" ").str[1].str.split("m").str[0]
```

```python
df["Duration_Minute"] = df["Duration_Minute"].fillna(0)
df[df["Duration_Minute"].isnull()]
```

```python
df.drop("Duration", axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
```

```python
encoder.fit_transform(
    df[["Airline", "Source", "Destination", "Additional_Info"]]
).toarray()
```
