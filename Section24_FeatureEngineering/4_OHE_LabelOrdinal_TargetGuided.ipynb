# Encoding

## Nominal/OHE(One Heat Encoding)
Categorical -> Numerical Values
- Uses Bits to Store categorical feature

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

df = pd.DataFrame({"color": ["red", "green", "blue", "red", "blue"]})
df
```

```python
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[["color"]]).toarray()
encoded
```

```python
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
encoded_df
```

```python
# Assignment
import seaborn as sns

eg_df = sns.load_dataset("tips")
eg_df
```

```python
encoder = OneHotEncoder()
encoded = encoder.fit_transform(
    eg_df[["sex", "smoker", "day", "time", "size"]]
).toarray()
encoded
```

```python
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
encoded_df
```

## Label & Ordinal Encoder
Label Encoding (Category -> numerical w/o Rank)
Ordinal Encoding (Category -> numerical w Rank)

```python
from sklearn.preprocessing import LabelEncoder
#Label Encoder
lbl_encoder = LabelEncoder()
encoder = lbl_encoder.fit_transform(df[["color"]])
encoder
```

```python
# Ordinal Encoder
from sklearn.preprocessing import OrdinalEncoder

ord_encoder = OrdinalEncoder(
    categories=[["school", "ug", "graduate", "pg"]]
)  # 2D Because Multiple columns rank
df = pd.DataFrame(
    {"points": ["school", "ug", "graduate", "pg", "ug", "school", "graduate"]}
)
encoded = ord_encoder.fit_transform(df[["points"]])
encoded
```

## Target Guided Ordinal Encoding

```python
import pandas as pd

df = pd.DataFrame(
    {"city": ["Delhi", "UP", "Delhi", "Bihar"], "price": [200, 300, 180, 400]}
)

mean_price = df.groupby("city")["price"].mean().to_dict()
mean_price
```

```python
df["city_encoded"] = df["city"].map(mean_price)
df[["city_encoded", "price"]]
```
